## <span style="color:#4682B4;">1.1. Introduction to Generative AI & Large Language Models (LLMs)</span>

### Generative AI
- Generative AI refers to a class of artificial intelligence systems designed to generate new content. This content can be text, images, audio, or even video. Unlike traditional AI systems that might classify or analyze existing data, generative AI creates something novel by learning patterns from a dataset and then producing similar but new data.

****
##### <span style="color:#008000;">Examples of Generative AI:</span>

- **Text Generation:** This includes systems that can write essays, articles, or even poetry. For instance, GPT-3 (the model behind me) can generate coherent and contextually relevant text based on a given prompt. For example, ChatGPT can write essays or answer questions based on what you ask.


- **Image Generation:** Models like DALL-E and Midjourney can create images from textual descriptions. For example, you could input "a futuristic cityscape at sunset" and the AI would generate an image that matches this description.


- **Music Composition:** Tools like OpenAI’s MuseNet can compose original music in various styles based on a given input, like generating a piece of music that combines elements of classical and jazz.

### Large Language Models (LLMs)

Large Language Models are a specific type of generative AI focused on understanding and generating human language. They are trained on vast amounts of text data to predict the next word in a sequence, allowing them to generate coherent and contextually appropriate sentences, paragraphs, or even entire articles.


****
##### <span style="color:#008000;">How They Work:</span>


- **Training:** LLMs are trained using a technique called unsupervised learning. They process a large corpus of text and learn to predict the likelihood of a word or phrase appearing in a given context. This involves analyzing the structure, grammar, and nuances of language.


- **Fine-tuning:** After the initial training, LLMs can be fine-tuned on specific datasets to improve performance on particular tasks or domains. For instance, a model might be fine-tuned to specialize in legal terminology for use in legal document analysis.


**Examples:**


- **ChatGPT:** This is a chatbot by OpenAI that can chat with you, write essays, answer questions, and more.


- **Google Bard:** Google’s AI that helps with generating text and answering questions, similar to ChatGPT.


- **Meta LLaMA 2:** A model by Meta (formerly Facebook) used for various text-based tasks, including generating responses and understanding text.


- **BERT (Bidirectional Encoder Representations from Transformers):** Developed by Google, BERT improves the understanding of the context of words in search queries, which helps in providing more relevant search results.

****
##### <span style="color:#008000;">Why It’s Important</span>


- **Enhanced Communication:** LLMs and generative AI can assist in breaking down language barriers by providing real-time translation or generating content in multiple languages. For example, businesses can use LLMs to translate their websites into various languages efficiently.


- **Content Creation:** These technologies enable the creation of new content at scale. For instance, marketing teams can use LLMs to generate social media posts, blog articles, or ad copy quickly and cost-effectively.


- **Personalization:** Generative AI can tailor content to individual preferences. For example, streaming services like Netflix use algorithms to recommend shows based on your viewing history, while LLMs can create personalized recommendations or messages.


- **Automation of Routine Tasks:** LLMs can automate repetitive tasks, such as drafting emails or summarizing documents. This can save time and reduce the need for manual intervention in routine workflows.


- **Creative Industries:** In fields like music, art, and literature, generative AI offers new tools for creativity. Artists can use AI to brainstorm ideas, generate drafts, or explore novel styles. For example, AI-generated art can provide inspiration for designers or help in creating unique visuals for advertising campaigns.


- **Assistance and Accessibility:** LLMs can provide real-time assistance through conversational agents (like chatbots), making information more accessible. They can also support individuals with disabilities by offering text-to-speech or speech-to-text services.

## <span style="color:#4682B4;">1.2. Basics of Prompt Engineering: Structuring Prompts to Get Desired Outputs</span>

### Prompt engineering
Prompt Engineering is the process of designing and crafting inputs (prompts) to get the best possible output from an AI model, especially those that generate text or perform language-based tasks. The goal is to structure your prompt in a way that guides the AI to produce useful, accurate, or creative responses.

**OR**

Prompt engineering is figuring out the best way to ask an AI a question or give it instructions so it gives you the best answer. It’s about crafting your requests to get useful and accurate information from the AI.

**Example:**
- Imagine you have a robot that can answer questions. If you ask it, `“Tell me about birds,”` it might give you a general answer. But if you ask, `“What are the unique features of parrots that help them mimic human speech?”` you’ll get a more specific and detailed answer about parrots.

> So, prompt engineering is about figuring out how to ask your questions or give instructions in a way that helps the AI give you the most useful and accurate information. It’s like fine-tuning your request to get exactly what you need!

****
##### <span style="color:#008000;">Why Prompt Engineering?</span>

Prompt engineering is crucial because the quality of the output from AI models like ChatGPT or GPT-4 depends heavily on how well the prompt is crafted. 

So, A well-designed prompt can:

- Improve the accuracy of the AI’s responses.
- Make the AI’s outputs more relevant and useful.
- Help achieve specific goals or gather particular types of information.


****
##### <span style="color:#008000;">Key Elements of a Prompt</span>

1. **Persona:** Defines the role or perspective that the AI should take.

    - **Example:** "As a career counselor, suggest five potential career paths for someone interested in technology."


2. **Instruction:** Provides clear guidance on what the AI should do.

    - **Example:** "List the top three benefits of regular exercise."


3. **Context:** Offers background information or sets the scene for the AI’s response.

    - **Example:** "In the context of a startup company with limited budget, outline cost-effective marketing strategies."


4. **Input Data:** Includes any specific details or data that the AI should consider while generating the response.

    - **Example:** "Given that the student scored 85% in math and 90% in science, suggest suitable high school electives."


5. **Output Indicator:** Specifies how the response should be formatted or what type of output is expected.

    - **Example:** "Provide a bulleted list of pros and cons."

****
##### <span style="color:#008000;">Examples for Students</span>

**1. Math Problem Explanation:**

- Prompt: "Explain how to solve quadratic equations using the quadratic formula, with an example problem and step-by-step solution."
- Persona: Math tutor
- Instruction: Explain solving quadratic equations
- Context: Educational tutorial
- Input Data: Example problem
- Output Indicator: Step-by-step solution

**2. Historical Research:**

- Prompt: "Summarize the main causes of the American Civil War, highlighting political, economic, and social factors. Use - bullet points for clarity."
- Persona: History professor
- Instruction: Summarize causes
- Context: American Civil War
- Input Data: Political, economic, social factors
- Output Indicator: Bullet points

****
##### <span style="color:#008000;">Techniques in Prompt Engineering</span>

1. **Specificity:** Be precise about what you want. Clear and specific prompts yield better results.

    - Technique: Use detailed instructions and provide clear examples.

2. **Iterative Refinement:** Start with a broad prompt and then refine based on the responses you get.

    - Technique: Adjust and experiment with different prompt phrasings.

3. **Contextual Information:** Provide enough background information to guide the AI’s responses.

    - Technique: Include relevant details and set the scene appropriately.

4. **Use of Examples:** Give examples of the kind of responses you’re expecting.

    - Technique: Show sample outputs to guide the AI.

****
##### <span style="color:#008000;">Pros and Cons of Prompt Engineering</span>


**Pros:**
- **Improved Accuracy:** Well-crafted prompts lead to more accurate and relevant responses from the AI.

- **Efficiency:** Helps get the desired information or output quickly without multiple iterations.

- **Customization:** Allows for tailored responses that fit specific needs or contexts.

**Cons:**
- **Learning Curve:** Crafting effective prompts can be challenging and requires practice.

- **Over-Reliance:** Over-engineering prompts might limit the AI’s ability to generate creative or varied responses.

- **Variability:** Different AI models might respond differently to the same prompt, requiring adjustments and experimentation.

## Key Types of Prompts in Prompt Engineering:

In prompt engineering, different types of prompts are used to guide AI models to generate specific types of outputs or handle various tasks effectively. 

- Here’s a detailed explanation of three key types of prompts: 

1. Zero-shot Prompting
2. Few-shot Prompting
3. Chain-of-Thought (CoT) Prompting

---
##### <span style="color:#008000;">1. Zero-shot prompting</span>

Zero-shot prompting means asking the AI to perform a task or answer a question without providing any examples. The AI relies on its general knowledge to respond. This technique is useful when you want the model to handle straightforward tasks based on its existing understanding.

- **Example:** “Describe the concept of blockchain technology.”

**Use Cases:**
- **Translation Tasks:** Asking the model to translate text between languages.
- **General Information Requests:** Inquiries where the model is expected to provide information or answer questions without additional context or examples.
- **Simple Classification:** For tasks like sentiment analysis where the model must infer the category based on its training.

---
##### <span style="color:#008000;">2. Few-Shot Prompting</span>

Few-shot prompting involves providing the model with a few examples of the desired input-output pairs before asking it to generate or handle similar tasks. This approach helps the model understand the task better by seeing specific examples of how to perform it. For example, if you want the AI to generate creative writing, you might provide several sample sentences to guide its responses.

- **Example:** “Here are a few examples of product reviews: [Example 1: The XYZ Phone is excellent for its battery life and camera quality.] [Example 2: The ABC Laptop is powerful with its high-speed processor and large screen.] Write a review for a new smartwatch.”

**Use Cases:**
- **Task-specific Applications:** When you need the model to perform a task with a certain format or style, such as specific types of translations or text formatting.
- **Improving Accuracy:** Providing examples can help the model generate more accurate responses by learning from the given examples.
- **Generating Content:** Creating content with specific patterns or styles based on provided examples.

---
##### <span style="color:#008000;">3. Chain-of-Thought (CoT) Prompting</span>

Chain-of-Thought (CoT) prompting involves guiding the model through a step-by-step reasoning process to arrive at the final answer. This type of prompting helps the model perform complex reasoning tasks by breaking down the problem into smaller, manageable steps.

**Example**
```python
To solve the following math problem, follow these steps:
Problem: "What is 15 divided by 3?"
Steps:
1. Identify the numbers involved in the problem.
2. Perform the division operation.
3. State the result.

Now solve: "What is 72 divided by 8?"
    
**Output**
1. The numbers involved are 72 and 8.
2. Performing the division: 72 ÷ 8 = 9.
3. The result is 9.



**Use Cases:**
- **Complex Reasoning Tasks:** When the problem requires multiple steps or logical reasoning, such as solving mathematical problems or performing detailed analysis.
- **Educational Tools:** Teaching or tutoring applications where step-by-step solutions are beneficial for learning.
- **Problem-Solving:** Any scenario where detailed explanations or logical steps are needed to achieve the final answer.

## <span style="color:#4682B4;">2.2. Introduction to Langchain: A Framework for Building AI-powered Apps</span>



****
##### <span style="color:#008000;">What is LangChain?</span>



LangChain is a powerful tool that can be used to work with Large Language Models (LLMs). LLMs are very general in nature, which means that while they can perform many tasks effectively, they may not be able to provide specific answers to questions or tasks that require deep domain knowledge or expertise. 

**For example:** Imagine you want to use an LLM to answer questions about a specific field, like medicine or law. While the LLM may be able to answer general questions about the field, it may not be able to provide more detailed or nuanced answers that require specialized knowledge or expertise.

> LangChain is a framework for developing applications powered by language models

**Why we need LangChain?**
LangChain offers a useful approach where the corpus of text is preprocessed by breaking it down into chunks or summaries, embedding them in a vector space, and searching for similar chunks when a question is asked. 

This pattern of preprocessing, real-time collecting, and interaction with the LLM is common and can be used in other scenarios as well, such as code and semantic search. 

LangChain provides an abstraction that simplifies the process of composing these pieces. This “prompt plumbing” is crucial and will become increasingly important as LLMs become more powerful and require more data to be provided at prompt time.


****
##### <span style="color:#008000;">LangChain Components</span>


**There are six main areas that LangChain is designed to help with.**

1. Model I/O
2. Data Connection
3. Chains
4. Agents
5. Memory
6. Collbacks


> The framework also allows integration with many tools to develop full-stack applications, such as OpenAI, Huggingface Transformers, and Vectors stores like Pinecone and chromadb, among others.


##### <span style="color:#008000;">Comprehensive Explanation of Components:</span>

1. **Model I/O:** Interface with language models. It consists of Prompts, Models, and Output parsers


2. **Data connection:** Interface with application-specific data sources with data transformers, text splitters, vector stores, and retrievers


3. **Chains:** Construct a sequence of calls with other components of the AI application. some examples of chains are sequential chains, summarization chain, and Retrieval Q&A Chains


4. **Agents:** LangChain introduces the notion of agents, which are entities that can perform specific tasks or actions. Agents can be used to automate interactions with LLMs, such as executing commands or retrieving information based on natural language inputs.


5. **Memory:** The framework supports memory management, enabling applications to maintain context across interactions. This is particularly useful for creating conversational agents that need to remember previous interactions.


6. **Callbacks:** Log and stream steps of sequential chains in order to run the chains efficiently and monitor the resources consumption

##### <span style="color:#008000;">LLMs</span>


There are two types of language models, which in LangChain are called:
    
**LLMs:** this is a language model which takes a string as input and returns a string

**ChatModels:** this is a language model which takes a list of messages as input and returns a message


The input/output for LLMs is simple and easy to understand - a string. But what about ChatModels? 

The input there is a list of **ChatMessages**, and the output is a single **ChatMessage**. A **ChatMessage** has two required components:

- **content:** This is the content of the message.

- **role:** This is the role of the entity from which the ChatMessage is coming from.

**LangChain provides several objects to easily distinguish between different roles:**

- **HumanMessage:** A ChatMessage coming from a human/user.
- **AIMessage:** A ChatMessage coming from an AI/assistant.
- **SystemMessage:** A ChatMessage coming from the system.
- **FunctionMessage:** A ChatMessage coming from a function call.

**Objective:** Build a LangChain-based application that generates common interview questions and answers for a specific job role.

**Task:**

- **Prompt Template:** Design a prompt that takes a job role (e.g., software engineer, marketing manager) and generates a list of common interview questions along with suggested answers.
- **Example Input:** "Software Engineer."

- **Example Output:**
    - **"1. Question:** What are your strengths and weaknesses? Answer: My strengths include problem-solving and attention to detail. My weakness is that I sometimes spend too much time on perfectionism, but I am working on managing my time better."

    - **"2. Question:** Can you describe a challenging project you worked on? Answer: I worked on a project to redesign our company's website. The challenge was integrating various APIs, but I overcame it by collaborating closely with the team and ensuring thorough testing."
    - **"3. Question:** How do you handle tight deadlines? Answer: I prioritize tasks based on urgency and importance, and I ensure regular communication with stakeholders to manage expectations and deliver on time."

````python
from langchain import PromptTemplate, LLMChain
import openai

# Set up your OpenAI API key
#openai.api_key = 'your-api-key-here'

# Define the template for generating interview questions and answers
template = """You are an HR specialist preparing for a job interview. Based on the job role of {job_role}, generate a list of 3 common interview questions along with suggested answers.
Format the response as follows:
1. Question: [Question] Answer: [Answer]
2. Question: [Question] Answer: [Answer]
3. Question: [Question] Answer: [Answer]"""

# Create the PromptTemplate object
interview_prompt = PromptTemplate(
    input_variables=["job_role"],
    template=template
)

# Format the prompt with a specific job role
formatted_prompt = interview_prompt.format(job_role='Software Engineer')

# Initialize the LLMChain with the prompt template and OpenAI function
llm_chain = LLMChain(
    llm=llm,
    prompt=interview_prompt
)

# Generate the response using the LLMChain
response = llm_chain.run(job_role='Software Engineer')

# Print the output
print(f"LLM Output: {response}")
````
